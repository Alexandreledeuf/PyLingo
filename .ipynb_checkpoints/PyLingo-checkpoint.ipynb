```python
import pandas as pd
import numpy as np
import re
import sys
import random
from random_word import RandomWords
from collections import Counter

def randomSizeWord():
    return random.randint(5, 10)

def userInput(lenWord):
    while True:
        word = input("\nEnter a word of size %s:"%lenWord)
        word = word.lower()
        if len(word) != lenWord:
            print("Error! Only %s characters allowed!"%lenWord)
        else:
            break
    letters_guess = list(word)
    print("\n Word enter: %s"%(letters_guess))
    return letters_guess


def randomWord():
    r = RandomWords() #Initialize
    lenWord = randomSizeWord() #Obtain random size of word
    randomWord = r.get_random_word(minLength=lenWord, maxLength=lenWord) # Return a single random word
    randomWord = randomWord.lower()
    
    return randomWord

def WordInformations(word):
    letters = list(word)
    lenWord = len(letters)
    letters_found = [" "]*lenWord
    Attempt = letters_found*6
    Guess = letters_found
    letters_found[0] = letters[0]
    print(letters_found)
    letter_count = Counter(letters)
    liste_occurence_letter = []
    return letters, lenWord, letters_found, Guess, letter_count, liste_occurence_letter

def getLetterInformations(letter_count,liste_occurence_letter):
    for k,v in letter_count.most_common():
        liste_occurence_letter.append(k)
        liste_occurence_letter.append(v)
    return liste_occurence_letter

def CorrectWrongLetter(liste_occurence_letter, letters, letters_guess, Guess):
    for k in range(0,len(liste_occurence_letter), 2):
        for j in range(len(letters)):
            for i in range(len(letters_guess)):
                if((letters_guess[i]==letters[j]) and (i==j) and (liste_occurence_letter[k]==letters_guess[i] and liste_occurence_letter[k+1]>0 and ((Guess[i] == " ") or (Guess[i] == "X")))): #
                    Guess[i] = "O"
                    liste_occurence_letter[k+1] -= 1 
                    break
                elif ((letters_guess[i]==letters[j]) and (i!=j) and (liste_occurence_letter[k]==letters_guess[i] and liste_occurence_letter[k+1]>0 and ((Guess[i] == " ")or (Guess[i] == "X")))): #
                    Guess[i] = "X"
                    liste_occurence_letter[k+1] -= 1 
                    break
                    
    print("\n Guess:%s"%(Guess))
    return liste_occurence_letter, Guess

def letterFoundDisplay(Guess, letters_found, letters):
    for i in range(1,len(Guess)):
        if(Guess[i]=="O"):
            letters_found[i]=letters[i]
        elif(Guess[i]==" " or Guess[i]=="X"):
            letters_found[i] = " "
        else:
            letters_found[i]=letters[i]
            
    print("\n Letters found %s"%(letters_found))
    return letters_found

def verificationWin(letters_found, word):
    GuessWord = "".join(letters_found)
    if(GuessWord==word):
        print("\n %s, You win !"%(word))
        return True
    else:
        return False

def main():
    word = randomWord()
    letters, lenWord, letters_found, Guess, letter_count, liste_occurence_letter = WordInformations(word)
    Attempt = 6
    while Attempt > 0:
        letters_guess = userInput(lenWord)
        liste_occurence_letter = getLetterInformations(letter_count,liste_occurence_letter)
        liste_occurence_letter, Guess = CorrectWrongLetter(liste_occurence_letter, letters, letters_guess, Guess)
        letters_found = letterFoundDisplay(Guess, letters_found, letters)    
        winCondition = verificationWin(letters_found, word)
        if(winCondition==True):
            break
        Attempt -= 1
    print("You lost ! the word was %s"%(word))

if __name__ == "__main__":
    main()
```
